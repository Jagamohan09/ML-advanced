<h1>Part 1</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cluster_mpg.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies = pd.get_dummies(df.drop('name',axis=1))
```

```python
df_w_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Using minmaxscaler 
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data,columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8),dpi=200)
sns.heatmap(scaled_df,cmap='viridis')
```

```python
plt.figure(figsize=(15,8),dpi=200)
sns.clustermap(scaled_df,row_cluster=False)
```

```python
plt.figure(figsize=(15,8),dpi=200)
sns.clustermap(scaled_df,col_cluster=False)
```

<h1>Part 2</h1>

```python
len(df)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4,metric='euclidean',linkage='ward')
```

```python
scaled_df
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='mpg',y='weight',hue=cluster_labels,palette='viridis')
```

**In MinMax Scaler the maximum distance possible is equal to the sqrt(no of features)**

```python
np.sqrt(len(scaled_df.columns))
# MAX DISTANCE POSSIBLE BETWEEN TWO POINTS
```

```python
# Distance_threshhold = 0 every datapoint is considered as to be a own cluster
model = AgglomerativeClustering(n_clusters=None,distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10),dpi=200)
dendro = dendrogram(linkage_matrix,truncate_mode='lastp',p=10)
```

```python
plt.figure(figsize=(20,10),dpi=200)
dendro = dendrogram(linkage_matrix,truncate_mode='level',p=3)
```
