```python
#we drop the species column becoz it is our target column
X = pd.get_dummies(df.drop('species',axis=1),dtype=float,drop_first=True)
```

```python
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

<h1>Part 2</h1>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay,confusion_matrix
```

```python
print(classification_report(y_test,base_pred))
```

```python
cm = confusion_matrix(y_test,base_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,
            columns=['Feature Importance']).sort_values('Feature Importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8),dpi=200)
plot_tree(model,filled=True,rounded=True,feature_names=list(X.columns));
```

**Creating a Function so that we don't have to write the above steps once again**

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=list(X.columns));
```

```python
# Reducing the Depth of the Tree
pruned_tree = DecisionTreeClassifier(max_depth=3)
```

```python
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
#determining the max_leaf_nodes that are allowed to be in the tree
max_leaf_node = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_node.fit(X_train,y_train)
```

```python
report_model(max_leaf_node)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
