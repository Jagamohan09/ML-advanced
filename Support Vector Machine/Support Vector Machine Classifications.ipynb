```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mouse_viral_study.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='Med_1_mL',y='Med_2_mL',hue='Virus Present',
                data=df,palette='seismic')

# HyperPlane for a 2D is a line

x = np.linspace(0,10,100)
m = -1
b = 11
y = m*x + b

plt.plot(x,y,'black')
```

**Lets use sklearn to Find the best hyperplane**

```python
from sklearn.svm import SVC
```

```python
#help(SVC)
```

```python
y = df['Virus Present']
```

```python
X = df.drop('Virus Present',axis=1)
```

```python
model = SVC(kernel = 'linear',C=1000)
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model,X,y)
```


<h1>Part 2</h1>

```python
model = SVC(kernel='linear',C=0.05)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C =1,gamma='auto')
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C =1,gamma='scale')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1,gamma=2)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly',degree=2,C=0.05)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.01,0.1,1],'kernel':['linear','rbf']}
```

```python
grid = GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.best_params_
```
